```python
# https://github.com/bmabey/pyLDAvis
# http://nbviewer.ipython.org/github/bmabey/pyLDAvis/blob/master/notebooks/pyLDAvis_overview.ipynb#topic=17&lambda=0.27&term=
# http://nbviewer.ipython.org/github/bmabey/pyLDAvis/blob/master/notebooks/Gensim%20Newsgroup.ipynb#topic=21&lambda=1&term=
# https://pyldavis.readthedocs.org/en/latest/readme.html

import gensim
import pyLDAvis.gensim
```

## 10,000 tweet sample

```python
dictionary = gensim.corpora.Dictionary.load('20150629_genocideDictGensim.dict')
corpus = gensim.corpora.MmCorpus('20150629_genocideCorpusGensim.mm')
lda = gensim.models.ldamodel.LdaModel.load('20150629_genocideLDAGensim.model')
```

```python
lda.show_topics()
```

```python
vis_data = pyLDAvis.gensim.prepare(lda, corpus, dictionary)
```

```python
pyLDAvis.display(vis_data)
```
